```python
same = "_same"

for dataset_name in ["adult", "bank", "heloc", "cancer", "wine"]:
    folder = os.path.join(os.getcwd(), "Neigh", dataset_name)
    folder_all = os.path.join(os.getcwd(), "Neigh", f"all{same}")
    folder_save = os.path.join(os.getcwd(), "Neigh", f"final{same}")

    idx_same = np.load(os.path.join(folder, "model_concordance_idx_test.npy"))
    disagree_dict = joblib.load(os.path.join(folder, "disagree_dict.joblib"))
    n = len(idx_same)

    print("\n", n)
    knn = disagree_dict["knn"]
    rf = disagree_dict["rf"]
    nn = disagree_dict["nn"]

    a = len(knn)
    b = len(rf)
    c = len(nn)
    print(a, b, c)

    n_neigh_agree = np.load(os.path.join(folder, f"n_neigh_expl_concordant{same}.npy"))
    n_neigh_disagree = np.load(os.path.join(folder, f"n_neigh_expl_disconcordant{same}.npy"))

    expl_agree = np.load(os.path.join(folder, f"expl_concordant{same}.npy"))
    expl_disagree = np.load(os.path.join(folder, f"expl_disconcordant{same}.npy"))


    agree = np.load(os.path.join(folder_all, f"{dataset_name}_aggregations.npy"))
    disagree = np.load(os.path.join(folder_all, f"{dataset_name}_aggregations_disagree.npy"))

    agree_rob = np.load(os.path.join(folder_all, f"{dataset_name}_robustness.npy"))
    disagree_rob = np.load(os.path.join(folder_all, f"{dataset_name}_robustness_disagree.npy"))


    idx_dict = {"agree": idx_same,
                "knn": knn, 
                "rf": rf, 
                "nn": nn}

    joblib.dump(idx_dict, os.path.join(folder_save, f"{dataset_name}_idx.joblib"))
    
    n_neigh = np.zeros(n_neigh_agree.shape)
    explanation = np.copy(expl_agree)
    aggregation = np.zeros(agree.shape)
    robustness = np.zeros(agree_rob.shape)

    n_neigh[idx_same] = n_neigh_agree[:n]
    #explanation[idx_same] = expl_agree[:n]
    aggregation[idx_same] = agree[:n]
    robustness[idx_same] = agree_rob[:n]

    n_neigh[knn] = n_neigh_disagree[:a]
    explanation[knn] = expl_disagree[:a]
    aggregation[knn] = disagree[:a]
    robustness[knn] = disagree_rob[:a]

    n_neigh[rf] = n_neigh_disagree[a:(a+b)]
    explanation[rf] = expl_disagree[a:(a+b)]
    aggregation[rf] = disagree[a:(a+b)]
    robustness[rf] = disagree_rob[a:(a+b)]

    n_neigh[nn] = n_neigh_disagree[(a+b):]
    explanation[nn] = expl_disagree[(a+b):]
    aggregation[nn] = disagree[(a+b):]
    robustness[nn] = disagree_rob[(a+b):]
    

    np.save(os.path.join(folder_save, f"{dataset_name}_n_neigh"), n_neigh)
    np.save(os.path.join(folder_save, f"{dataset_name}_explanation"), explanation)
    np.save(os.path.join(folder_save, f"{dataset_name}_aggregation"), aggregation)
    np.save(os.path.join(folder_save, f"{dataset_name}_robustness"), robustness)
```
